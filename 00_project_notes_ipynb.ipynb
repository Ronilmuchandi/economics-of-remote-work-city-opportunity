# The Economics of Remote Work-City-Level Opportunity & Fragility (2022–2025)

## Scope
- Global cities (50–100)
- Timeframe: 2022–2025
- Focus: Business + Finance
- Goal: Opportunity Index and Fragility Index

## Roadmap
1. Select cities
2. Select datasets
3. Clean and merge data
4. Build indices
5. Analyze and model
6. Scenario analysis
7. Dashboard and insights

## Primary Dataset (Spine)

Remote job demand by US city over time (2022–2025).

This dataset represents the core signal of the project and captures how demand
for remote roles is distributed and evolving across major US cities.
All other datasets are used to explain, stress-test, or contextualize this signal.
